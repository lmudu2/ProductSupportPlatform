```python
#@title 2. Configure API Keys
import os
from google.colab import userdata

try:
    groq_key = userdata.get('GROQ_API_KEY')
    os.environ["GROQ_API_KEY"] = groq_key
    with open('.env', 'w') as f:
        f.write(f"GROQ_API_KEY={groq_key}")
    print("✅ Groq API Key loaded from Colab Secrets.")
except:
    print("⚠️ GROQ_API_KEY not found in Secrets. Please add it to avoid AI errors.")
```

```python
#@title 3. Launch Dashboard with Secure Native Proxy
import os
import time
import socket
from google.colab import output

def wait_for_port(port, timeout=40):
    start_time = time.time()
    while True:
        try:
            with socket.create_connection(("127.0.0.1", port), timeout=1):
                return True
        except (OSError, ConnectionRefusedError):
            if time.time() - start_time > timeout:
                return False
            time.sleep(1)

print("🧹 Cleaning up existing processes...")
!fuser -k 8501/tcp

print("🚀 Launching Product Support Intelligence Dashboard...")
# Stability flags for Colab iframe environments
get_ipython().system_raw("streamlit run main.py --server.port 8501 --server.address 0.0.0.0 --server.enableCORS False --server.enableXsrfProtection False --server.enableWebsocketCompression False > streamlit.log 2>&1 &")

if wait_for_port(8501):
    print("✅ Dashboard Backend Ready!")
    print("\n--- 🔗 SECURE DASHBOARD LINK ---")
    
    # Native Google Secure Proxy (No 'Connection not private' errors, No localtunnel crashes)
    try:
        proxy_url = output.eval_js("google.colab.kernel.proxyPort(8501)")
        print(f"Dashboard URL: {proxy_url}")
        print("--------------------------------\n")
        print("Access Instructions:")
        print("1. Click the secure link above.")
        print("2. It will open safely in a new tab without any security warnings.")
        print("   (If asked to confirm leaving Google, click 'Continue').")
    except Exception as e:
        print("Error generating URL. Please ensure you are connected to a Colab runtime.")
else:
    print("❌ Startup failed. Displaying logs:")
    !cat streamlit.log
```
